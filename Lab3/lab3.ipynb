Mozemy to zrobic w nieco ladniejszy sposob za pomoca funkcji.

```python
from typing import Dict

def Encode(Train_set: pd.Series, Test_set: pd.Series) -> Dict:
    if Train_set.empty or Test_set.empty:
        raise ValueError("Empty set passed!")

    TrainTest = pd.concat([Train_set, Test_set])
    return {j: i for i, j in enumerate(TrainTest.unique())}
```

Sprawdzmy, czy funkcja dziala poprawnie.

```python
dict_test = Encode(X_train['CabinReduced'], X_test['CabinReduced'])
dict(itertools.islice(dict_test.items(), 10))
```

<h1> Zadanie 7 </h1>
Tworzymy liste slownikow dla calego zbioru X.

```python
dictionaries_list = [Encode(X_train[name], X_test[name]) for name in col_name]
```

Mapujemy, najpierw dla X_train.

```python
X_train_mapped=[X_train[col].map(dictionaries_list[i]) for i, col in enumerate(X_train.columns)]
X_train_mapped[0]
```

Dla ladniejszego wyswietlania mozemy zrobic:

```python
X_train_mapped = pd.concat(X_train_mapped, axis = 1)
X_train_mapped
```

Analogicznie postepujemy dla X_test.

```python
X_test_mapped=[X_test[col].map(dictionaries_list[i]) for i, col in enumerate(X_test.columns)]
X_test_mapped=pd.concat(X_test_mapped, axis=1)
X_test_mapped
```
